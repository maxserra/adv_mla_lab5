```python
import pandas as pd
```

```python
data = pd.read_csv("../data/raw/german_credit.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt

plt.hist(data["Creditability"])
plt.show()
```

## Prepare the data

```python
y = data.pop("Creditability")
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data,
                                                    y,
                                                    test_size=0.2,
                                                    random_state=23)
```

```python
print(f"X_train {X_train.shape}")
print(f"y_train {y_train.shape}")
print(f"X_test  {X_test.shape}")
print(f"y_test  {y_test.shape}")
```

```python
from sklearn.preprocessing import scale
```

```python
X_train = scale(X_train)
X_test = scale(X_test)
```

```python
X_train[0]
```

## Define the model

```python
import numpy as np

import tensorflow as tf


tf.random.set_seed(23)
np.random.seed(23)
```

```python
regularizer = tf.keras.regularizers.l1_l2()
```

```python
model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Dense(128, activation="relu", kernel_regularizer=regularizer,
                                input_shape=(20, )))
model.add(tf.keras.layers.Dropout(rate=0.25))
model.add(tf.keras.layers.Dense(1, activation="sigmoid"))

model.compile(loss=tf.keras.losses.binary_crossentropy,
              optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.005),
              # optimizer=tf.keras.optimizers.Adam(learning_rate=0.005),
              metrics=[tf.keras.metrics.AUC(name="auc"),
                       # tf.keras.metrics.binary_accuracy,
                      ])
```

```python
model.summary()
```

## Train the model

```python
epochs = 250

history = model.fit(X_train,
                    y_train,
                    batch_size=None,
                    epochs=epochs,
                    verbose=2,
                    validation_split=0.2)
```

```python
model.evaluate(X_test, y_test)
```

```python
epochs = range(1, len(history.history["auc"])+1)

plt.plot(epochs, history.history["auc"],
         label='Training auc')
plt.plot(epochs, history.history["val_auc"],
         label='Validation auc')
plt.xlabel('Epochs')
plt.ylabel('auc')
plt.legend()

plt.show()
```
